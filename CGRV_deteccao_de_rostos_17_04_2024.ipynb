```python
from google.colab import files
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

# Função para carregar e exibir a imagem
def carregar_e_exibir_imagem():
    # Carregar a imagem do computador
    uploaded = files.upload()

    # Iterar sobre as chaves do dicionário uploaded
    for filename, data in uploaded.items():
        # Verificar se o arquivo é uma imagem
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            # Ler a imagem carregada
            img = cv2.imdecode(np.frombuffer(data, np.uint8), cv2.IMREAD_COLOR)

            # Mostrar a imagem
            cv2_imshow(img)
            return img

    # Se nenhum arquivo de imagem for encontrado, exibir uma mensagem de erro
    print("Erro: Nenhuma imagem carregada.")
    return None

# Função para detectar e desenhar retângulos em torno das faces na imagem
def detectar_faces(img):
    # Transformar a imagem em tons de cinza
    imgPB = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Carregar o classificador de faces
    #df = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    #df = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_fullbody.xml')
    #df = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_upperbody.xml')
    df = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_lefteye_2splits.xml')


    # Utilizar a função de detecção a partir do objeto df
    faces = df.detectMultiScale(imgPB, scaleFactor=1.05, minNeighbors=7, minSize=(30,30), flags=cv2.CASCADE_SCALE_IMAGE)

    # Desenhar retângulos amarelos nas posições onde foram encontradas as faces
    for(x,y,w,h) in faces:
        cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,255), 7)

    # Mostrar a imagem atualizada com as marcações das faces
    cv2_imshow(img)

# Carregar e exibir a imagem
img = carregar_e_exibir_imagem()

# Verificar se a imagem foi carregada corretamente antes de continuar
if img is not None:
    # Detectar e desenhar retângulos em torno das faces na imagem
    detectar_faces(img)
```
