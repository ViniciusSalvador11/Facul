```python
import cv2 as cv
from google.colab.patches import cv2_imshow
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread('/content/partenon.jpg')
assert img is not None, "file could not be read, check with os.path.exists()"
#cv2_imshow(img)

#BORRAMENTO DE IMAGEM

blur = cv2.blur(img, (5,5))

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Borrada')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Filtro de Passa-alta
import cv2 as cv
from google.colab.patches import cv2_imshow
import numpy as np
from matplotlib import pyplot as plt

imagem_cinza = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv2_imshow(imagem_cinza)
kernel = np.array([[-1, -1, -1],
                  [-1, 9, -1],
                  [-1, -1, -1]])
filtro = cv.filter2D(imagem_cinza, -1, kernel)
cv2_imshow(filtro)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
